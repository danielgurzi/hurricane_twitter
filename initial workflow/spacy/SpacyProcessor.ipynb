```python
import pandas as pd
import json
import spacy


```

```python
nlp = spacy.load("harvey_model")
```

```python
df = pd.read_csv("../Project_5/data/harvey/hurricaneharvey_50000.csv")
```

```python
df['text'].dropna(inplace = True)
```

```python
text_list = {}

for i in range(10000):
    text = df['text'][i]

    # Process the text
    doc = nlp(text)

    # Iterate over the entities
    for ent in doc.ents:
        # Print the entity text and label
#         print(ent.text, ent.label_)
        if ent.label_:
            text_list.update({i : [ent.text, doc.text]})
            
```
